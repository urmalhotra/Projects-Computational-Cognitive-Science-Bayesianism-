```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import math

mean = 0
sigma_squared = 1
x1 = 2
sigma_squared_1 = 0.25
sigma_squared_2 = 4

posterior_mean_1 = ((mean*(1/sigma_squared) + (1/sigma_squared_1)*x1)/
((1/sigma_squared) + 1*(1/sigma_squared_1)))
posterior_sigma_squared_1 = 1/(1/(sigma_squared) + 1*(1/sigma_squared_1))

predictive_mean_1 = posterior_mean_1
predictive_sigma_squared_1 = posterior_sigma_squared_1 + sigma_squared_1

posterior_mean_2 = ((mean*(1/sigma_squared)) + (x1/sigma_squared_2))/((1/sigma_squared) + (1/sigma_squared_2))
posterior_sigma_squared_2 = 1/((1/sigma_squared) + (1/sigma_squared_2))

predictive_mean_2 = posterior_mean_2
predictive_sigma_squared_2 = posterior_sigma_squared_2 + sigma_squared_2

fig, ax = plt.subplots(ncols = 2, figsize= (12,5))

x_posterior = np.linspace(-5,7,200)
x_predictive = np.linspace(-7,8,200)
ax[0].plot(x_posterior, norm.pdf(x_posterior,posterior_mean_1,math.sqrt(posterior_sigma_squared_1)), color = 'red')
ax[0].plot(x_posterior, norm.pdf(x_posterior,posterior_mean_2,math.sqrt(posterior_sigma_squared_2)), color = 'blue')
ax[0].set_xlabel("\u03BC")
ax[0].set_ylabel("p(\u03BC)")
ax[0].legend(["$\u03C3_x = 0.25$", "$\u03C3_x = 4$"])
ax[0].set_title("Posterior Distribution")

ax[1].plot(x_predictive, norm.pdf(x_predictive,predictive_mean_1,math.sqrt(predictive_sigma_squared_1)), color = 'red')
ax[1].plot(x_predictive, norm.pdf(x_predictive,predictive_mean_2,math.sqrt(predictive_sigma_squared_2)), color = 'blue')
ax[1].set_xlabel("$x_{N+1}$")
ax[1].set_ylabel("p(\u03BC)")
ax[1].legend(["$\u03C3_x = 0.25$", "$\u03C3_x = 4$"])
ax[1].set_title("Predictive Distribution")

print(posterior_mean_1)
print(posterior_mean_2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import math

mean = 0
sigma_squared = 1
x1 = [2.1,2.5,1.4,2.2,1.8]
sigma_squared_1 = 0.25
sigma_squared_2 = 4

posterior_mean_1 = ((mean*(1/sigma_squared) + (1/sigma_squared_1)*sum(x1))/
((1/sigma_squared) + 5*(1/(sigma_squared_1))))
posterior_sigma_squared_1 = 1/(1/(sigma_squared) + 5*(1/(sigma_squared_1)))

predictive_mean_1 = posterior_mean_1
predictive_sigma_squared_1 = posterior_sigma_squared_1 + sigma_squared_1

posterior_mean_2 = ((mean*(1/sigma_squared) + (1/sigma_squared_2)*sum(x1))/
((1/sigma_squared) + 5*(1/(sigma_squared_2))))
posterior_sigma_squared_2 = 1/(1/(sigma_squared) + 5*(1/(sigma_squared_2)))

predictive_mean_2 = posterior_mean_2
predictive_sigma_squared_2 = posterior_sigma_squared_2 + sigma_squared_2

fig, ax = plt.subplots(ncols = 2, figsize= (12,5))

x_posterior = np.linspace(-2,4,300)
x_predictive = np.linspace(-6.5, 9, 300)
ax[0].plot(x_posterior, norm.pdf(x_posterior,posterior_mean_1,math.sqrt(posterior_sigma_squared_1)), color = 'red')
ax[0].plot(x_posterior, norm.pdf(x_posterior,posterior_mean_2,math.sqrt(posterior_sigma_squared_2)), color = 'blue')
ax[0].set_xlabel("\u03BC")
ax[0].set_ylabel("p(\u03BC)")
ax[0].legend(["$\u03C3_x = 0.25$", "$\u03C3_x = 4$"])
ax[0].set_title("Posterior Distribution")

ax[1].plot(x_predictive, norm.pdf(x_predictive,predictive_mean_1,math.sqrt(predictive_sigma_squared_1)), color = 'red')
ax[1].plot(x_predictive, norm.pdf(x_predictive,predictive_mean_2,math.sqrt(predictive_sigma_squared_2)), color = 'blue')
ax[1].set_xlabel("$x_{N+1}$")
ax[1].set_ylabel("p(\u03BC)")
ax[1].legend(["$\u03C3_x = 0.25$", "$\u03C3_x = 4$"])
ax[1].set_title("Predictive Distribution")

print(posterior_mean_1)
print(posterior_mean_2)
plt.show()
```
